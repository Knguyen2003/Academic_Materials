#### Import thư viện

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate 5000 numbers from a uniform distribution , setting the parameters low to -3 and high to 3
low = -3
high = 3
uniform_numbers = np.random.uniform(low, high, 5000)
#.uniform : Phương thức này được sử dụng để tạo ra các số ngẫu nhiên từ một phân phối đều (uniform distribution) trong một khoảng cho trước.
#Hiển thị 10 số đầu
print(uniform_numbers[:10])
```

```python
# Generate 5000 numbers from a normal  distribution , setting the parameters loc to 5 and scale to 2
loc = 5
scale = 2
normal_numbers = np.random.normal(loc, scale, 5000)
#.normal là Phương thức được sử dụng để tạo ra các số ngẫu nhiên từ một phân phối chuẩn (normal distribution), còn được gọi là phân phối Gaussian.
#Phương thức này sẽ tạo ra các số ngẫu nhiên từ phân phối chuẩn với trung bình loc và độ lệch chuẩn scale.
#Trong phân phối chuẩn, phần lớn các giá trị sẽ tập trung xung quanh giá trị trung bình và giảm dần khi di chuyển ra xa hai bên từ giá trị trung bình.

#Hiển thị 10 số đầu
print(normal_numbers[:10])
```

```python
#Plot a histogram of uniforms with bins of width of 0.25 from -3 to 3 using plt.hist()

# Tạo 5000 số ngẫu nhiên từ phân phối đều từ -3 đến 3
uniform_numbers = np.random.uniform(-3, 3, 5000)

# Xác định các ranh giới bin với độ rộng là 0.25
bin_edges = np.arange(-3, 3.25, 0.25) #3.25 là giá trị kết thúc của phạm vi.Lưu ý không bao gồm 3.25, vì vậy bin cuối cùng sẽ kết thúc ở 3.

# Vẽ biểu đồ histogram
plt.hist(uniform_numbers, bins=bin_edges, edgecolor='black')

# Thêm nhãn và tiêu đề
plt.xlabel('Giá trị')
plt.ylabel('Tần suất')
plt.title('Biểu đồ histogram của phân phối đều \n')

# Hiển thị biểu đồ
plt.show()
```

```python
#Plot a histogram of normals with bins of width of 0.5 from -2 to 13 using plt.hist()

# Tạo 5000 số từ phân phối chuẩn với trung bình 5 và độ lệch chuẩn 2
normal_numbers = np.random.normal(5, 2, 5000)

# Xác định các ranh giới bin với độ rộng là 0.5 từ -2 đến 13
bin_edges = np.arange(-2, 13.5, 0.5)

# Vẽ biểu đồ histogram
plt.hist(normal_numbers, bins=bin_edges, edgecolor='black')

# Thêm nhãn và tiêu đề
plt.xlabel('Giá trị')
plt.ylabel('Tần suất')
plt.title('Biểu đồ histogram của phân phối chuẩn')

# Hiển thị biểu đồ
plt.show()
```
