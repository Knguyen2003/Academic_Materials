#### Bài tập: Cho các số a, b, m, n trong đó a < b là hai số thực bất kỳ; m, n là các số nguyên dương. Viết hàm số tạo một mảng hai chiều có shape = (m, n), các phần tử là các số ngẫu nhiên phân bố đều trong nửa đoạn [a, b).

```python
a = 1
b = 5
m = 3
n = 4
random_array = np.random.uniform(a, b, (m, n))
#Hiển thị mảng hai chiều với các phần tử là các số ngẫu nhiên phân bố đều
print(random_array)
```

```python
#Kiểm tra mảng trả về có đúng là mảng ngẫu nhiên các phần tử trong nửa đoạn [a,b)

#Tính kỳ vọng (mean) và phương sai (variance) của mảng
mean_array = np.mean(random_array)
variance_array = np.var(random_array)
print("Kỳ vọng của mảng: ", mean_array)
print("Phương sai của mảng: ", variance_array)

expected_mean = (b + a) / 2
expected_variance = ((b - a) ** 2) / 12
print("Kỳ vọng của biến ngẫu nhiên theo phân phối đều: ", expected_mean)
print("Phương sai của biến ngẫu nhiên theo phân phối đều: ", expected_variance)
#Nếu mảng được tạo ra chính xác theo phân phối đều trong nửa đoạn [a, b), thì kỳ vọng và phương sai của mảng sẽ
#gần bằng với kỳ vọng và phương sai của biến ngẫu nhiên theo phân phối đều.
```

#### Bài tập: Cho các số a, s, m, n với:

* a là một số thực bất kỳ.
* s là một số thực dương.
* m, n là các số nguyên dương.
#### Xây dựng một mảng ngẫu nhiên hai chiều có shape = (m, n) mà các phần tử của nó tuần theo phân phối chuẩn có kỳ vọng bằng a và phương sai là s.

```python
a = 5.2  # Kỳ vọng
s = 6.7  # Phương sai
m = 5  # Số hàng
n = 4  # Số cột

#Tạo mảng
array_random = np.random.normal(a, np.sqrt(s), (m, n))
#Hàm np.random.normal(a, np.sqrt(s), size) tạo ra một mảng các số ngẫu nhiên tuân theo phân phối chuẩn với kỳ vọng là a và độ
#lệch chuẩn là căn bậc hai của phương sai (vì np.random.normal() yêu cầu đầu vào là độ lệch chuẩn chứ không phải phương sai).
print(array_random)
```
