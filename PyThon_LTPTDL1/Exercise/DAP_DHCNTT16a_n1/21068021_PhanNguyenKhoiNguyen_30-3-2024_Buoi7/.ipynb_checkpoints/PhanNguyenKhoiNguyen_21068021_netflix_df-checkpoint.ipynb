#### Đọc dữ liệu

1. Create a DataFrame, netflix_df from the CSV provided, available at the
path "netflix_data.csv" and print the first five rows of the DataFrame to inspect the data and ensure it was created successfully.

```python
netflix_df  = pd.read_csv(r"T:\data\netflix_data.csv")
netflix_df.head(5)
```

2. Subset the DataFrame for "Movies" 
- Subset the netflix_df DataFrame such that only rows where the type is a "Movie" are preserved. Assign the result to netflix_subset.

```python
netflix_subset = netflix_df[netflix_df['type'] == 'Movie']
netflix_subset.head()
```

3. Subset the columns of the new DataFrame
- Subset the netflix_subset DataFrame again to preserve only the columns title, country, genre, release_year, and duration and save the DataFrame as netflix_movies DataFrame.

```python
# Chỉ định các cột cần giữ lại
columns_to_keep = ['title', 'country', 'genre', 'release_year', 'duration']

# Chọn một phần của DataFrame để chỉ giữ lại các cột đã chỉ định
netflix_movies = netflix_subset.loc[:, columns_to_keep].copy()
#.loc là một phương thức cho phép chọn một phần của DataFrame theo các nhãn hoặc điều kiện. Trong trường hợp này, chúng ta sử 
#dụng .loc để chọn tất cả các hàng (:) và chỉ các cột được chỉ định trong danh sách columns_to_keep

# Hiển thị kết quả
netflix_movies.head()
```

4. Filter the DataFrame by Movie duration
- Filter netflix_movies to see movies that have a duration fewer than 60 minutes, save as short_movies and print the first 20 rows to inspect the data.

```python
# Lọc các bộ phim có thời lượng ít hơn 60 phút từ netflix_movies
short_movies = netflix_movies[netflix_movies['duration'] < 60]

# In ra 20 dòng đầu tiên để kiểm tra dữ liệu
print(short_movies.head(20))
```

5. Assign colors to movie genres
Initialize an empty list called colors to store your different color values and using a for loop, iterate through netflix movies rows and append colors to your list based on the genres "Children", "Documentaries", "Stand-Up", and other.

```python
# Khởi tạo một danh sách rỗng để lưu trữ các giá trị màu
colors = []

# Lặp qua các dòng của phim Netflix
for index, row in netflix_movies.iterrows():
    genre = row['genre']
    if 'Children' in genre:
        colors.append('blue')  # Gán màu xanh cho thể loại Trẻ em
    elif 'Documentaries' in genre:
        colors.append('green')  # Gán màu xanh lá cây cho thể loại Tài liệu
    elif 'Stand-Up' in genre:
        colors.append('red')  # Gán màu đỏ cho thể loại Hài Stand-Up
    else:
        colors.append('gray')  # Gán màu xám cho các thể loại khác
        
# print(colors)
```

6. Plot the movie duration by year using the genre colors created
Using netflix_movies, initialize a figure object called fig with width 12 and height a to create a scatter plot with release_year on the x-axis labelled "Release year", the movie duration on the y-axis labelled "Duration (min) ", the title "Movie Duration by Year of Release", and coloring the points with the new colors list.


```python
import matplotlib.pyplot as plt

# Initialize figure object
fig, ax = plt.subplots(figsize=(12, 9))  # Set width to 12 and height to a (specify a suitable value for 'a')

# Create scatter plot
scatter = ax.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors)

# Add labels and title
ax.set_xlabel('Release year')
ax.set_ylabel('Duration (min)')
ax.set_title('Movie Duration by Year of Release')

# Show the plot
plt.show()
```
