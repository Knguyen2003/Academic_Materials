#### Ex2:
Set the sample size to 70.
Calculate the population size from attrition_pop.
Calculate the interval between the rows to be sampled.Systematically sample attrition_pop to get the rows of the population at each interval, starting at 0; assign the rows to attrition_sys_samp.

```python
#Gán kích thước cho mẫu chuẩn bị lấy là 70
sample_size = 70
#Lấy kích thước của quần thể
population_size = len(attrition_pop)
# Khoảng cách giữa các hàng được lấy mẫu được tính bằng cách chia kích thước quần thể cho kích thước mẫu. 
#Điều này làm cho mỗi lần lấy mẫu, chúng ta lấy một hàng và bỏ qua các hàng tiếp theo theo khoảng cách tính được.
interval = population_size // sample_size
# .iloc[::interval], nó lấy mỗi hàng tại vị trí bắt đầu từ chỉ số 0 và bước sang hàng tiếp theo với khoảng cách là interval cho đến hết quần thể.
attrition_sys_samp = attrition_pop.iloc[::interval]
#Hiển thị mẫu sau khi lấy
print(attrition_sys_samp)
```

#### Ex3:	Get the proportion of employees by Education level from attrition_pop.
* Use proportional stratified sampling on attrition_pop to sample 40% of each Education group, setting the seed to 2022.
* Get the proportion of employees by Education level from attrition_strat.

```python
# Tính tỷ lệ nhân viên theo trình độ học vấn trong attrition_pop
education_proportions = attrition_pop['Education'].value_counts(normalize=True) #normalize=True, kết quả sẽ trả về tỷ lệ tần suất của mỗi giá trị.

# Lấy mẫu theo tỷ lệ dựa trên từng nhóm trình độ học vấn
sampled_data = attrition_pop.groupby('Education', group_keys=False).apply(lambda x: x.sample(frac=0.4, random_state=2022))
#attrition_pop.groupby('Education', group_keys=False): nhóm dữ liệu trong attrition_pop theo cột 'Education'. Điều này tạo ra một đối tượng DataFrameGroupBy, trong đó mỗi nhóm tương ứng với một giá trị trong cột 'Education'.
# .apply(lambda x: x.sample(frac=0.4, random_state=2022)): hàm sample() lên mỗi nhómđể lấy một mẫu ngẫu nhiên từ mỗi nhóm với tỷ lệ frac=0.4, nghĩa là lấy ra 40% của mỗi nhóm. 
# Tham số random_state=2022 được sử dụng để đảm bảo kết quả có thể tái tạo. Cuối cùng, chúng ta gom nhóm các mẫu này lại thành một DataFrame duy nhất, và kết quả là sampled_data.

# Tính tỷ lệ nhân viên theo trình độ học vấn trong tập dữ liệu đã lấy mẫu
education_proportions_strat = sampled_data['Education'].value_counts(normalize=True)

# In ra tỷ lệ nhân viên theo trình độ học vấn trong attrition_pop
print("Tỷ lệ nhân viên theo trình độ học vấn trong attrition_pop:")
print(education_proportions)

# In ra tỷ lệ nhân viên theo trình độ học vấn trong tập dữ liệu đã lấy mẫu
print("\nTỷ lệ nhân viên theo trình độ học vấn trong tập dữ liệu đã lấy mẫu:")
print(education_proportions_strat)
```

#### ex4: 
Create a list of unique JobRole values from attrition_pop, and assign to job_roles_pop
Randomly sample four JobRole values from job_roles_pop.
Subset attrition_pop for the sampled job roles by filtering for rows where JobRole is in job_roles_samp.
Remove any unused categories from JobRole.
For each job role in the filtered dataset, take a random sample of ten rows, setting the seed to 2022.

```python
# Tạo một danh sách các giá trị JobRole duy nhất
job_roles_pop = attrition_pop['JobRole'].unique()

# Lấy mẫu ngẫu nhiên bốn giá trị JobRole của job_roles_pop
job_roles_samp = np.random.choice(job_roles_pop, size=4, replace=False) 
#Tham số replace=False đảm bảo rằng không có giá trị nào được lặp lại trong kết quả.

# Lọc tập dữ liệu attrition_pop cho các vai trò công việc đã lấy mẫu
attrition_samp = attrition_pop[attrition_pop['JobRole'].isin(job_roles_samp)].copy()
#.isin(job_roles_samp) để lọc các hàng trong attrition_pop trong đó cột "JobRole" nằm trong danh sách các giá trị đã lấy mẫu job_roles_samp.
# .copy() được sử dụng để tạo một bản sao của DataFrame được lọc.

# Loại bỏ bất kỳ loại không sử dụng nào từ cột JobRole
attrition_samp['JobRole'] = pd.Categorical(attrition_samp['JobRole'], categories=job_roles_samp)
#Sử dụng pd.Categorical() để chuyển cột "JobRole" thành một categorical dtype, trong đó chỉ có các giá trị đã lấy mẫu.
#Điều này đảm bảo rằng chỉ có các loại đã lấy mẫu được giữ lại, các loại không sử dụng khác sẽ bị loại bỏ.

# Đối với mỗi vai trò công việc trong tập dữ liệu đã lọc, lấy một mẫu ngẫu nhiên của mười hàng
sampled_data = pd.DataFrame()
for role in job_roles_samp:
    role_data = attrition_samp[attrition_samp['JobRole'] == role].sample(n=10, random_state=2022)
    sampled_data = pd.concat([sampled_data, role_data]) ##pd.concat :nối 2 DataFrame
# Với mỗi giá trị của job_roles_samp, chúng ta lấy mẫu 10 hàng từ attrition_samp trong đó "JobRole" trùng khớp với vai trò công việc đó.
# ham số random_state=2022 đảm bảo kết quả có thể tái tạo.
# In ra tập dữ liệu đã lấy mẫu
print(sampled_data)
```
