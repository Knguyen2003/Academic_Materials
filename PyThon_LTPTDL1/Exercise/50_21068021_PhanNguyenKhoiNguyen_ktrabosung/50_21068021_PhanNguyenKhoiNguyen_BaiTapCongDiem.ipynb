Import thư viện

```python
import numpy as np
import pandas as pd
```

Đọc dữ liệu

```python
crime = pd.read_csv(r'/content/crimes.csv')
crime.head()
```

1. Giờ nào có tần suất tội phạm cao nhất

```python
# Chuyển cột 'TIME OCC' sang định dạng datetime
crime['TIME OCC'] = pd.to_datetime(crime['TIME OCC'])

# Tạo một cột mới 'Hour' để lấy giờ từ thời gian
crime['Hour'] = crime['TIME OCC'].dt.hour

# Đếm số lượng tội phạm theo giờ
crime_counts_by_hour = crime['Hour'].value_counts()

# Tìm giờ có tần suất tội phạm cao nhất
hour_with_highest_crime = crime_counts_by_hour.idxmax()
highest_crime_frequency = crime_counts_by_hour.max()

# Hiển thị kết quả
print("Giờ có tần suất tội phạm cao nhất là {}:00 với {} vụ.".format(hour_with_highest_crime, highest_crime_frequency))
```

2. Khu vực nào có tần suất tội phạm ban đêm cao nhất (Tội phạm xảy ra trong khoảng thời gian từ 22h đến 3h59 sáng)

```python
# Chuyển cột 'TIME OCC' sang định dạng datetime
crime['TIME OCC'] = pd.to_datetime(crime['TIME OCC'])

# Tạo một cột mới 'Hour' để lấy giờ từ thời gian
crime['Hour'] = crime['TIME OCC'].dt.hour

# Lọc dữ liệu để chỉ giữ lại các báo cáo tội phạm ban đêm (từ 22h đến 3h59)
night_crime_data = crime[(crime['Hour'] >= 22) | (crime['Hour'] < 4)]

# Đếm số lượng tội phạm theo khu vực
crime_counts_by_area = night_crime_data['AREA NAME'].value_counts()

# Tìm khu vực có tần suất tội phạm cao nhất
area_with_highest_crime = crime_counts_by_area.idxmax()
highest_crime_frequency = crime_counts_by_area.max()

# Hiển thị kết quả
print("Khu vực có tần suất tội phạm ban đêm cao nhất là '{}' với {} vụ xảy ra trong khoảng thời gian từ 22h đến 3h59 sáng.".format(area_with_highest_crime, highest_crime_frequency))
```

3.  Xác định số lượng tội ác được thực hiện đối với nạn nhân ở các nhóm tuổi khác nhau ("0-17", "18-25", "26-34", "35-44", "45-54", "55-64" và "65+")

```python
# Tạo một cột mới 'Victim Age Group' để phân loại nạn nhân theo nhóm tuổi
bins = [0, 17, 25, 34, 44, 54, 64, float('inf')]
labels = ['0-17', '18-25', '26-34', '35-44', '45-54', '55-64', '65+']
crime['Victim Age Group'] = pd.cut(crime['Vict Age'], bins=bins, labels=labels, right=False)

# Đếm số lượng tội ác theo nhóm tuổi của nạn nhân
crime_counts_by_age_group = crime['Victim Age Group'].value_counts()

# Hiển thị kết quả
print("Số lượng tội ác được thực hiện đối với nạn nhân ở các nhóm tuổi khác nhau:")
print(crime_counts_by_age_group)
```
