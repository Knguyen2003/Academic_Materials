Import thư viện

```python
import numpy as np
import pandas as pd
```

Đọc dữ liệu

```python
nobel = pd.read_csv(r'/content/nobel.csv')
nobel.head()
```

1 .  Nam hay nữ đạt giải Nobel nhiều hơn

```python
#print(nobel['sex'].unique())
#Đếm số lượng Nam và Nữ đạt giải Nobel
soLuong = nobel['sex'].value_counts()
#Hiển thị kết quả
print("Số lượng Nam đạt giả Nobel: ",soLuong.get('Male',0))
print("Số lượng Nữ đạt giả Nobel: ",soLuong.get('Female',0))

if soLuong.get('Male',0) > soLuong.get('Female',0):
  print("Nam đạt giải Nobel nhiều hơn")
elif soLuong.get('Male',0) < soLuong.get('Female',0):
  print("Nữ đạt giải Nobel nhiều hơn")
else:
  print("Nam và Nữ đạt giải Nobel bằng nhau")
```

2. Danh sách 5 quốc gia đạt giải Nobel nhiều nhất

```python
#Đếm số lượng đạt giải Nobel theo quốc gia
soLuongQuocGia = nobel['birth_country'].value_counts()
# print("Top 5 quốc gia đạt giải Nobel nhiều nhất: ")
#Hiển thị 5 quốc gia đạt giải Nobel nhiều nhất
print("Top 5 quốc gia đạt giải Nobel nhiều nhất: ")
soLuongQuocGia.head(5)
```

3. Thập kỷ nào mà tỉ lệ người đạt giải nobel sinh ra ở Mỹ là lớn nhất

```python
#Tạo cột decade
nobel["decade"] = (nobel['year'] // 10)
# Lọc những người đạt giải nobel sinh ra ở mỹ
nobel_My = nobel[(nobel["birth_country"] == 'USA') |(nobel["birth_country"] == 'United States of America')]
# Dem số lượng người đạt giải nobel theo thập kỷ (chỉ người Mỹ)
nguoiMy_Nobel_ThapKy = nobel_My['decade'].value_counts().sort_index()
# Đếm tổng số lượng người đạt giải nobel theo thập kỷ (toàn bộ người)
tongSLNguoi_ThapKy = nobel['decade'].value_counts().sort_index()
# Tính tỷ lệ người đạt giải nobel sinh ra ở mỹ theo thập kỷ
tiLe = nguoiMy_Nobel_ThapKy / tongSLNguoi_ThapKy
# Tìm thập kỷ có tỉ lệ người đạt giải nobel sinh ra ở mỹ là lớn nhất
thapKy_TiLeLonNhat = tiLe.idxmax()
tiLeLonNhat = tiLe.max()
print("Thập kỷ có tỉ lệ người đạt giải nobel sinh ra ở Mỹ là lớn nhất: ", thapKy_TiLeLonNhat)
print("Tỷ lệ: ", tiLeLonNhat)
```

4. Ai là người phụ nữ đầu tiên nhận giải Nobel và ở hạng mục nào

```python
#Lọc ra những người là phụ nữ
nobel_female =nobel[nobel['sex'] == 'Female']
# nobel_female.head(10)
#Sắp xếp năm giảm dần và tìm người đầu tiên
nobel_sorted_year_des = nobel_female.sort_values(by = 'year', ascending=True).iloc[0]
#Hiển thị kết quả
print("Người phụ nữ đầu tiên nhận giải Nobel: ")
print("Tên: ", nobel_sorted_year_des['full_name'])
print("Năm: ", nobel_sorted_year_des['year'])
print("Hạng mục: ", nobel_sorted_year_des['category'])
```

5. Những cá nhân hoặc tổ chức nào đã giành được nhiều hơn một giải thưởng nobel trong suốt những năm qua

```python
# Ghép cột "full_name" và cột "organization_name" thành Series
ten = nobel['full_name']
toChuc = nobel['organization_name']
tenVaToChuc = pd.concat([ten, toChuc])
# Đếm số lượng giải nobel mà mỗi người hoặc tổ chức đã giành được
demSL_Nobel = tenVaToChuc.value_counts()
# Lọc những người hoặc tổ chức có nhiều hơn 1 giải nobel
locNguoiCoNhieuHonMot = demSL_Nobel[demSL_Nobel > 1]
#Hiển thị kết quả
print("Những cá nhân hoặc tổ chức đã giành được nhiều hơn một giải thưởng nobel trong suốt những năm qua: ")
print(locNguoiCoNhieuHonMot)
```
